```python
#results that I want to return in each time step
    
def number_infected(model):
    l=[]
    for s in model.strain_list:
        l.append(sum([1 if s in a.infection_list else 0 for a in model.schedule.agents])) 
    return l

def total_number_infected(model):
    
    l=sum([1 if len(a.infection_list) > 0 else 0 for a in model.schedule.agents]) 
    return l

def number_susceptible(model):
    l=[]
    for s in model.strain_list:
        l.append(sum([1 if a.amISusceptible(s)==1 else 0 for a in model.schedule.agents]))      
    return l

def number_immune(model):
    l=[]
    for s in model.strain_list:
        l.append(sum([1 if a.amIImmune(s)==1 else 0 for a in model.schedule.agents]))      
    return l

def total_pop(model):
    l=sum([1 for a in model.schedule.agents])
    return l

def calculate_reff(model):
    reff_current=[np.mean(list(itertools.chain.from_iterable(model.moving_reff_dict[key]))) for key in model.moving_reff_dict]
    return reff_current
    
def calculate_reff_at(model):
    if model.no_loci_AT > 0:
        #display("Time %s"%model.time)
        #display(model.moving_reff_at_dict)
        reff_at_current=[np.mean(list(itertools.chain.from_iterable(model.moving_reff_at_dict[key]))) for key in model.moving_reff_at_dict]
        #display(model.moving_reff_at_dict)
        #display(reff_at_current)
        return reff_at_current
    
def calculate_reff_pathogen(model):
    #display(model.moving_reff_pathogen["ReffPathogen"])
    reff_current=np.mean(list(itertools.chain.from_iterable(model.moving_reff_pathogen["ReffPathogen"])))#np.mean(model.reff_pathogen) 
    return reff_current
def calculate_weekly_reff(model):
    reff_current=[np.mean(model.reff_dict[key]) for key in model.reff_dict]
    return reff_current
    
def calculate_weekly_reff_at(model):
    if model.no_loci_AT > 0:
        reff_at_current=[np.mean(model.reff_at_dict[key]) for key in model.reff_at_dict]
        return reff_at_current
    
def calculate_weekly_reff_pathogen(model):
    reff_current=np.mean(model.reff_pathogen) 
    return reff_current
def simpson_rec(model): 
    # Simpson reciprocal diversity index
    n_list=[]
    for s in model.strain_list:
        n_list.append(sum([1 if s in a.infection_list else 0 for a in model.schedule.agents])) 
    N=sum(n_list)
    #f is boolean to show that there is at least a value greater than 1 in the nlist
    f=sum([1 for k in n_list if k > 1 ])

    if f >= 1:
        n_list=np.array(n_list)  #n
        n1=n_list-np.array([1]*len(n_list))  #n-1
        D=(N*(N-1))/sum(n_list*n1)   #inverse of simpsons index
        return D
    else:
        return 0

def simpson_rec_at(model):
    if model.no_loci_AT > 0:
        last_infections=[k[-1] for k in model.ulist_at]
        #f is boolean to show that there is at least a value greater than 1 in the nlist
        f=sum([1 for k in last_infections if k > 1 ])
        if f >= 1:
            D= (sum(last_infections)*(sum(last_infections)-1))/sum([k*(k-1) for k in last_infections]) 
            return D
        else:
            return 0
    else:
        return 0
def simpson_rec_mt(model):
    if model.no_loci_MT > 0:
        last_infections=[k[-1] for k in model.ulist_mt]
        #f is boolean to show that there is at least a value greater than 1 in the nlist
        f=sum([1 for k in last_infections if k > 1 ])
        if f >= 1:
            D= (sum(last_infections)*(sum(last_infections)-1))/sum([k*(k-1) for k in last_infections]) 
            return D
        else:
            return 0
    else: 
        return 0

def cum_no_reintroduction_strains(model):
    return model.no_reintroduction_of_strains
    
def cum_no_reintroduction_ats(model):
    return model.no_reintroduction_of_ats

def cum_no_introduction_strains(model):
    return model.total_no_introduction_of_strains
    
def cum_no_introduction_ats(model):
    return model.total_no_introduction_of_ats

def duration_in_community_strain(model):
    duration_in_community=[np.nanmean(model.duration_in_community_strain_dict[key]) 
                           for key in model.duration_in_community_strain_dict]
    return duration_in_community

def duration_in_community_at(model):
    duration_in_community=[np.nanmean(model.duration_in_community_at_dict[key]) 
                           for key in model.duration_in_community_at_dict]
    return duration_in_community
def period_diversity(model):
    #period diversity from Bec's model
    total_infections=[sum(k) for k in model.ulist]
    #f is boolean to show that there is at least a value greater than 1 in the nlist
    f=sum([1 for k in total_infections if k > 1 ])
    if f >= 1:
        D= (sum(total_infections)*(sum(total_infections)-1))/sum([k*(k-1) for k in total_infections]) 
        return D
    else:
        return 0
def period_diversity_at(model):
    #period diversity from Bec's model
    if model.no_loci_AT >0:
        total_infections=[sum(k) for k in model.ulist_at]
        #display("model.ulist_at: %s"%model.ulist_at)
        #f is boolean to show that there is at least a value greater than 1 in the nlist
        f=sum([1 for k in total_infections if k > 1 ])
        if f >= 1:
            D= (sum(total_infections)*(sum(total_infections)-1))/sum([k*(k-1) for k in total_infections]) 
            return D
        else:
            return 0
    else:
        return 0
def period_diversity_mt(model):
    #period diversity from Bec's model
    if model.no_loci_MT >0:
        total_infections=[sum(k) for k in model.ulist_mt]
        #display("model.ulist_mt")
        #display(model.ulist_mt)
        #display("total_infections")
        #display(total_infections)
        #f is to show that there is at least a value greater than 1 in the nlist
        f=sum([1 for k in total_infections if k > 1 ])
        #display("f")
        #display(f)
        if f >= 1:
            D= (sum(total_infections)*(sum(total_infections)-1))/sum([k*(k-1) for k in total_infections]) 
            return D
        else:
            return 0
    else:
        return 0


def diversity_index(model):
    #known as Shannonâ€“Weaver diversity index
    #from Buckee 2004 and Shannon, C. E. & Weaver, W. (1949)
    #D ranges between 0 and 1, with D=1 indicating that allofthe possible 
    #strain types in the population are equally represented.
    # pi=is the frequency of strain i in the population
    n_list=[]
    for s in model.strain_list:
        n_list.append(sum([1 if s in a.infection_list else 0 
                           for a in model.schedule.agents])) 
    N=sum(n_list)
    
    if N > 0:  #if there is at least one infection
        n_list=[e for  e in n_list if e != 0]
        n_list=np.array(n_list) 
        p_list=n_list/([N]*len(n_list))
        D=sum(p_list*np.log(1/p_list))/np.log(N)
        return D
    else:
        return 0

def circulating_no_strains(model):
    n_list=[]
    for s in model.strain_list:
        n_list.append(sum([1 if s in a.infection_list else 0 
                           for a in model.schedule.agents])) 
    N=sum([1 for num in n_list if num != 0])
    return N

def hamming_distance_AT(model):
    if model.no_loci_AT >0:
        #from Buckee 2004
        wpp=[]
        pp=[]
        n_list=[]
        for s in model.strain_list:
            n_list.append(sum([1 if s in a.infection_list else 0 
                               for a in model.schedule.agents])) 
        N=sum(n_list)
        if N > 0:  #if there is at least one infection
            for j in np.arange(1,len(model.strain_list)):
                p_j=sum([1 if model.strain_list[j] in a.infection_list else 0 
                         for a in model.schedule.agents])/N
                for i in np.arange(0,j):  #i<j
                    p_i=sum([1 if model.strain_list[i] in a.infection_list else 0 
                             for a in model.schedule.agents])/N
                    w=0     #wij=the number of loci with different alleles for strains i and j
                    for h in range(0,model.no_loci_AT):
                        s1=model.strain_list[j][h]  #hth loci of strain 1
                        s2=model.strain_list[i][h]  #hth loci of strain 2
                        if s1 == s2:
                            w+=1
                    wpp.append(w*p_j*p_i)
                    pp.append(p_j*p_i)
            if sum(pp) > 0:
                H=(1/model.no_loci_AT)*sum(wpp)/sum(pp)
                return H
            else:
                return 0 #fix 
        else:
            return 0     
    else:
        return 0
    

def hamming_distance_MT(model):
    if model.no_loci_MT >0:
        #from Buckee 2004
        wpp=[]
        pp=[]
        n_list=[]
        for s in model.strain_list:
            n_list.append(sum([1 if s in a.infection_list else 0 
                               for a in model.schedule.agents])) 
        N=sum(n_list)

        if N > 0:  #if there is at least one infection
            for j in np.arange(1,len(model.strain_list)):
                p_j=sum([1 if model.strain_list[j] in a.infection_list else 0 
                         for a in model.schedule.agents])/N
                for i in np.arange(0,j):  #i<j
                    p_i=sum([1 if model.strain_list[i] in a.infection_list else 0 
                             for a in model.schedule.agents])/N
                    w=0     
                    #wij=the number of loci with different alleles for strains i and j
                    for h in range(model.no_loci_AT, 
                               model.no_loci_AT+model.no_loci_MT):
                        s1=model.strain_list[j][h]  #hth loci of strain 1
                        s2=model.strain_list[i][h]  #hth loci of strain 2
                        if s1 == s2:
                            w+=1
                    wpp.append(w*p_j*p_i)
                    pp.append(p_j*p_i)
            if sum(pp) > 0:
                H=(1/model.no_loci_MT)*sum(wpp)/sum(pp)
                return H
            else:
                return 0 #fix
        else:
            return 0
    else:
        return 0

def mean_cum_infections(model):
    l=np.mean([a.cumnoinfections for a in model.schedule.agents]) 
    return l
    
def circulating_no_ATs(model):
    if model.no_loci_AT >0: #if there is at least one AT locus
        
        n_list=[]
        for s in model.strain_list:
            #no of infection from each strains
            n_list.append(sum([1 if s in a.infection_list else 0 
                               for a in model.schedule.agents])) 
        N=sum(n_list)

        if N > 0:  #if there is at least one infection
            #circulating strain list
            clist=[model.strain_list[i] for i,num in enumerate(n_list) if num != 0]
            #AT values list from strain list
            clist_at=[clist[k][0:model.no_loci_AT] for k,l in enumerate(clist)]
            clist_at.sort()
            #number of at
            n_at=len(list(clist_at for clist_at,_ in itertools.groupby(clist_at))) 
            return n_at
        else: #if there isn't any infections
            return 0
    else: #if there isn't any AT locus
        return 0
    
def circulating_no_MTs(model):
    if model.no_loci_MT >0: #if there is at least one MT locus
        
        n_list=[]
        for s in model.strain_list:
            #no of infection from each strains
            n_list.append(sum([1 if s in a.infection_list else 0 
                               for a in model.schedule.agents])) 
        N=sum(n_list)

        if N > 0:  #if there is at least one infection
            #circulating strain list
            clist=[model.strain_list[i] for i,num in enumerate(n_list) if num != 0]
            #AT values list from strain list
            clist_mt=[clist[k][model.no_loci_AT:(model.no_loci_AT+model.no_loci_MT)] 
                      for k,l in enumerate(clist)]
            clist_mt.sort()
            #number of mt
            n_mt=len(list(clist_mt for clist_mt,_ in itertools.groupby(clist_mt))) 
            return n_mt
        else: #if there isn't any infections
            return 0
    else: #if there isn't any MT locus
        return 0
```
